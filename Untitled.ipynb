```python
days = ['MONDAY', 'TUESDAY', 'WEDNESDAY', 'THURSDAY', 'FRIDAY', 'SATURDAY', 'SUNDAY']
meal_types = ['BREAKFAST','LUNCH','SNACKS','DINNER']
menu = {
 
}
for day in range(len(days)):
    col = df.columns.get_loc(days[day])
    menu[days[day]] = {}
    row = df[df.iloc[:, 0] == meal_types[0] ].index[0]
    menu[days[day]][meal_types[0]] = {
        "Main Dish": [df.iloc[row,col],df.iloc[row+1,col] if not pd.isna(df.iloc[row + 1, col]) !='nan'else"" ],
            "Egg / Cereals": df.iloc[row+2,col].replace('+', ''),
            "Sprouts": df.iloc[row+3,col].replace('+', ''),
            "Tea and Coffee": df.iloc[row+4,col].replace('+', ''),
            "Fruit": df.iloc[row+5,col].replace('+', ''),
            "Milk": df.iloc[row+6,col].replace('+', '')
    }
    row = df[df.iloc[:, 0] == meal_types[1] ].index[0]
    menu[days[day]][meal_types[1]] ={
        "Dry":df.iloc[row,col] ,
            "Gravy": df.iloc[row+1,col],
            "Sp Rice": df.iloc[row+2,col],
            "India Bread": df.iloc[row+3,col],
            "Pickle": df.iloc[row+4,col],
            "Papad": df.iloc[row+5,col],
            "Salad": df.iloc[row+6,col],
            "Plain Rice": df.iloc[row+7,col],
            "Extra": df.iloc[row+8,col]
        
    }
    row = df[df.iloc[:, 0] == meal_types[2] ].index[0]
    menu[days[day]][meal_types[2]] ={
        "Snack": df.iloc[row,col],
            "Tea and Coffee": df.iloc[row+1,col],
            "Fruits": df.iloc[row+2,col]
    }
    row = df[df.iloc[:, 0] == meal_types[3] ].index[0]
    menu[days[day]][meal_types[3]] ={
            "Dry Vegetable": df.iloc[row,col],
            "Curry": df.iloc[row+1,col],
            "Salad": df.iloc[row+2,col],
            "Roti": df.iloc[row+3,col],
            "Pickle": df.iloc[row+4,col],
            "Plain Rice": df.iloc[row+5,col],
            "Soup": df.iloc[row+6,col],
            "Dessert": df.iloc[row+7,col],
            "Milk": df.iloc[row+8,col],
            "Extra": df.iloc[row+9,col]
        }
    
        
```

```python
import base64
from email.mime.text import MIMEText
from google_auth_oauthlib.flow import InstalledAppFlow
from googleapiclient.discovery import build
from requests import HTTPError

SCOPES = [
        "https://www.googleapis.com/auth/gmail.send"
    ]
flow = InstalledAppFlow.from_client_secrets_file('token.json', SCOPES)
creds = flow.run_local_server(port=0)
print("this is cred",creds)
service = build('gmail', 'v1', credentials=creds)
message = MIMEText('This is the body of the email')
message['to'] = 'movienight578@gmail.com'
message['subject'] = 'Test Mail'
create_message = {'raw': base64.urlsafe_b64encode(message.as_bytes()).decode()}

try:
    message = (service.users().messages().send(userId="me", body=create_message).execute())
    print(F'sent message to {message} Message Id: {message["id"]}')
except HTTPError as error:
    print(F'An error occurred: {error}')
    message = None
```
